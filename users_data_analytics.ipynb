```python
#get payments csv file data
payments_data=pd.read_csv("payments.csv",index_col=0)
payments_data
```

```python
#combine orders and payments table based on shared order_id
orders_and_payments_table=pd.merge(
    orders_data,
    payments_data,
    on="order_id",
    how="outer")
orders_and_payments_table
```

```python
#combine users with orders and payments table
users_orders_payments_table=pd.merge(
    users_data,
    orders_and_payments_table,
    on="user_id",
    how="outer")
#reset index
users_orders_payments_table=(
    users_orders_payments_table
    .reset_index(drop=True)
    .assign(user_id=lambda x:x.index+1)
    .set_index("user_id")
)
users_orders_payments_table
```

```python
#check for missing values in users_orders_payments_table
#find total count for each missing value
users_orders_payments_table.isnull().sum()
```
